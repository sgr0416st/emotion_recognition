```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
def draw_line_graph_from_csv(file_name, x_axis, y_axes=None):
    df = pd.read_csv(file_name)
    filter = [x_axis] + y_axes

    title = os.path.basename(file_name)
    df[filter].plot(x=x_axis, title=title, grid=True, rot=10, figsize=(6, 4))
    csv_dir = '/Users/satousuguru/workspace/programing/python/emotion_recognition/samples/recognition_results/'
```

```python
csv_dir = '/Users/satousuguru/workspace/programing/python/emotion_recognition' \
          '/samples/datas/emotion_record_sample_data/'
draw_line_graph_from_csv(
    csv_dir + 'test_change.csv'
    , 'second'
    , ['anger', 'contempt', 'disgust', 'fear', 'happiness', 'neutral', 'sadness', 'surprise']
    )
```

```python
csv_dir = '/Users/satousuguru/workspace/programing/python/emotion_recognition' \
          '/samples/datas/emotion_record_sample_data/'
csv_name = (
    'with_sota_kagimoto_cut_slice_change.csv',
    'with_sota_kobayashi_cut_slice_change.csv',
    'with_sota_zyo_cut_slice_change.csv'
)

for name in csv_name:
    draw_line_graph_from_csv(
        csv_dir + name
        , 'second'
        , ['anger', 'contempt', 'disgust', 'fear', 'happiness', 'neutral', 'sadness', 'surprise']
        )
```
